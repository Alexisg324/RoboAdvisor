# Compile and Evaluate a Linear Regression Model Using a Neural Network

### Step 1: Create a deep neural network by assigning the number of input features, the number of layers, and the number of neurons on each layer using Tensorflowâs Keras.

```python
# Define the the number of inputs (features) to the model
number_input_features = 57

# Review the number of features
number_input_features
```

```python
# Define the number of neurons in the output layer
number_output_neurons = 1
number_output_neurons
```

```python
# Define the number of hidden nodes for the first hidden layer
hidden_nodes_layer1 = round((number_input_features+number_output_neurons)/2)
# Review the number hidden nodes in the first layer
hidden_nodes_layer1
```

```python
# Define the number of hidden nodes for the second hidden layer
hidden_nodes_layer2 =  round((hidden_nodes_layer1+number_output_neurons)/2)
# Review the number hidden nodes in the second layer
hidden_nodes_layer2
```

```python
# Create the Sequential model instance
nn = Sequential()
```

```python
# Add the first hidden layer
# YOUR CODE HERE
nn.add(Dense(
    units = hidden_nodes_layer1,
    activation = 'relu',
    input_dim = number_input_features
))
```

```python
nn.add(Dense(
    units = hidden_nodes_layer1,
    activation = 'relu',
))
```

```python
# Add the first hidden layer
# YOUR CODE HERE
nn.add(Dense(
    units = hidden_nodes_layer2,
    activation = 'relu'
))
```

```python
# Add the first hidden layer
# YOUR CODE HERE
nn.add(Dense(
    units = hidden_nodes_layer2,
    activation = 'relu'
))
```

```python
# Add the output layer to the model specifying the number of output neurons and activation function
# YOUR CODE HERE
nn.add(Dense(units=1, activation ='linear'))
```

```python
# Display the Sequential model summary
# YOUR CODE HERE
nn.summary()
```

### Step 2: Compile and fit the model using the `binary_crossentropy` loss function, the `adam` optimizer, and the `accuracy` evaluation metric.

```python
# Compile the Sequential model
# YOUR CODE HERE
nn.compile(optimizer='adam',loss='mse',metrics='mean_absolute_error')
```

```python
# Fit the model using 50 epochs and the training data
# YOUR CODE HERE
nn.fit(x_train_lr, y_train_lr, epochs=500)
```

### Step 3: Evaluate the model using the test data to determine the modelâs loss and accuracy.

```python
# Evaluate the model loss and accuracy metrics using the evaluate method and the test data
mean_absolute_error = nn.evaluate(x_test_lr,y_test_lr,verbose=2)

# Display the model loss and accuracy results
print(f"MSE: {mean_absolute_error }")
```
