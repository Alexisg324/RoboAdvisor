```python
#Create an empty results dictionary
result_dict = {}
```

```python
#Takes inputs from Build a model - y_test, y_pred - and calculates the rations
## This function is called from the model function 

def summarize_classification(y_test,
                            y_pred
                            ):
    acc = accuracy_score (y_test, y_pred, normalize = True)
    num_acc = accuracy_score (y_test, y_pred, normalize = False)
    prec = precision_score (y_test, y_pred)
    recall = recall_score (y_test, y_pred)
    
    return {
        'accuracy':acc,
        'precision': prec,
        'recall':recall,
        'accuracy_count':num_acc}   
```

```python
#Build a model that can work with several of the SKLearn functions
#1. Build Model -
## a.Inputs for the model will be the 
###a.1 classifier function (linerar regression, linear discriminant etc.,)
###b. name of y columns
###c. names of X columns
###d. dataset that includes x & y columns
###e. test fractions used to split the database into test and train sets - used in the train_test_split
## b.Identifies the X&Y columns in the dataset and the names of X columns and y columns specified  as input. 
## c.Peforms train test split on the dataset provided
## d.Peforms the instatntiation and training (fit) for the model that will be supplied later
## e.Peforms the prediction based on the test model 
## e.Peforms the prediction based on the test model 
## f.Calculate different ratios to identify the model efficiencies by calling the summarize_classification function defined earlier 
## g.Load the prediction results in a dataframe for analyzing the results.

def build_model(classifier_fn,
                name_of_y_col,
                names_of_x_cols,
                dataset,
                test_frac=0.2):
    X = dataset[names_of_x_cols]
    y = dataset[name_of_y_col]
    
    x_train, x_test, y_train, y_test = train_test_split(X,y, test_size = test_frac)
    
    model = classifier_fn (x_train, y_train)
    
    y_pred = model.predict(x_test)
    y_pred_train = model.predict(x_train)
    
    train_summary = summarize_classification (y_train, y_pred_train) 
    test_summary = summarize_classification (y_test, y_pred)
    
    pred_results = pd.DataFrame({'y_test':y_test,
                                'y_pred':y_pred})
    model_crosstab = pd.crosstab(pred_results.y_pred, pred_results.y_test)
    
    return {
        'training': train_summary,
        'test': test_summary,
        'confusion_matrix': model_crosstab
    }
```

```python
#Takes the results from the return and prints it out for comparision purposes
def compare_results():
    for key in result_dict:
        print('Classification:' , key)
        print()
        print('Training data')
        for score in result_dict[key]['training']:
            print(score, result_dict[key]['training'][score])
        print()
        print('Test data')
        for score in result_dict[key]['test']:
            print(score, result_dict[key]['test'][score])
        print()
```

```python
# RandomForestClassifier
def random_forest_fn(x_train, y_train):
    model = RandomForestClassifier()
    model.fit(x_train, y_train)
    return model
```

```python
# The final result analysis
# The sequence of the call is as below:
##1.Call compare resuls function
##2.This will call the results dictionary
##3.This will call the Build model function and calls the function definition
##4.Build Model function takes the details, summarizes the results, and prints out the resulkts dictionary
result_dict['stock-price-RandomForestClassifier'] = build_model(random_forest_fn,
                                                   'Target',
                                                   FEATURES,
                                                   stock_analysis_data)
compare_results()
```

```python
# This function will be called  in the build model and this is for Logistic Regression
def logistic_fn(x_train, y_train):
    model = LogisticRegression(solver = 'liblinear')
    model.fit(x_train, y_train)
    return model
```

```python
#The final result analysis
# The sequence of the call is as below:
##1.Call compare resuls function
##2.This will call the results dictionary
##3.This will call the Build model function and calls the function definition
##4.Build Model function takes the details, summarizes the results, and prints out the resulkts dictionary
result_dict['stock-price-LogisticRegression'] = build_model(logistic_fn,
                                                 'Target',
                                                   FEATURES,
                                                   stock_analysis_data)
compare_results()
```

```python
# This function will be called  in the build model and this is for Quadratic Discriminat Analysis 
def quadratic_discriminant_fn(x_train, y_train):
    model = QuadraticDiscriminantAnalysis()
    model.fit(x_train, y_train)
    return model
```

```python
# The final result analysis
# The sequence of the call is as below:
##1.Call compare resuls function
##2.This will call the results dictionary
##3.This will call the Build model function and calls the function definition
##4.Build Model function takes the details, summarizes the results, and prints out the resulkts dictionary
result_dict['stock-price -quadratic_discriminant_analysis'] = build_model(quadratic_discriminant_fn,
                                                                          'Target',
                                                                          FEATURES,
                                                                          stock_analysis_data)
compare_results()
```

```python
#Takes inputs from Build a model - y_test, y_pred - and calculates the rations
## This function is called from the model function 

def summarize_classification(ylr_test,
                            ylr_pred
                            ):
    r21_score = r2_score(ylr_test, ylr_pred, multioutput='variance_weighted')
    return {'R2_Score': r21_score}   
```

```python
#Build a model that can work with several of the SKLearn functions
#1. Build Model -
## a.Inputs for the model will be the 
###a.1 classifier function (linerar regression, linear discriminant etc.,)
###b. name of y columns
###c. names of X columns
###d. dataset that includes x & y columns
###e. test fractions used to split the database into test and train sets - used in the train_test_split
## b.Identifies the X&Y columns in the dataset and the names of X columns and y columns specified  as input. 
## c.Peforms train test split on the dataset provided
## d.Peforms the instatntiation and training (fit) for the model that will be supplied later
## e.Peforms the prediction based on the test model 
## e.Peforms the prediction based on the test model 
## f.Calculate different ratios to identify the model efficiencies by calling the summarize_classification function defined earlier 
## g.Load the prediction results in a dataframe for analyzing the results.

def build_model_lr(classifier_fn,
                name_of_ylr_col,
                names_of_xlr_cols,
                dataset,
                test_frac=0.2):
    Xlr = dataset[names_of_xlr_cols]
    ylr = dataset[name_of_ylr_col]
    
    xlr_train, xlr_test, ylr_train, ylr_test = train_test_split(Xlr,ylr, test_size = test_frac)
    
    model_lr = classifier_fn (xlr_train, ylr_train)
    
    ylr_pred = model_lr.predict(xlr_test)
    ylr_pred_train = model_lr.predict(xlr_train)
    
    train_summary = summarize_classification (ylr_train, ylr_pred_train) 
    test_summary = summarize_classification (ylr_test, ylr_pred)
    
    pred_results = pd.DataFrame({'ylr_test':ylr_test,
                                'ylr_pred':ylr_pred})
    
    return {
        'training': train_summary,
        'test': test_summary,
    }
```

```python
#Takes the results from the return and prints it out for comparision purposes
def compare_results():
    for key in result_dict:
        print('Classification:', key)
        print()
        print('Training data')
        for score in result_dict[key]['training']:
            print(score, result_dict[key]['training'][score])
        print()
        print('Test data')
        for score in result_dict[key]['test']:
            print(score, result_dict[key]['test'][score])
        print()
```

```python
#The final result analysis
# The sequence of the call is as below:
##1.Call compare resuls function
##2.This will call the results dictionary
##3.This will call the Build model function and calls the function definition
##4.Build Model function takes the details, summarizes the results, and prints out the results dictionary
result_dict['stock-price-LinearRegression'] = build_model_lr(linear_regression_fn,
                                                   'value',
                                                   FEATURES_lr,
                                                   stock_analysis_data_lr)
compare_results()
```
